```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sympy as sp
import networkx as nx
import os
%matplotlib inline
```

```python
ruta="/home/diego/Documentos/PiedrasTDA/"
```

```python
#Carga los archivos de exel en una lista con sus nombres
lst_arch=([x for x in os.listdir(ruta) if x.startswith('Matriz')])
lst_arch.remove('MatrizAdyacencia_MuestraDeMano.xlsx')
lista=[]
for i in lst_arch:
    df=pd.read_excel(ruta+i,header=None)
    lista.append((df,i))
len(lista)
```

```python
#Imprime el grado de cada nodo en la grafo
def grado(dflista):
    n=0
    grad=[]
    print(dflista[1])
    for i in range(len(dflista[0])):
        print(n,dflista[0][i].sum())
        grad.append(dflista[0][i].sum())
        n+=1
    return(grad)

#Regresa una lista de grafos con sus nombres
def graphs(lista):
    grafs=[]
    for i in lista:
        M=i[0].as_matrix()
        G=nx.Graph(M)
        grafs.append([G,i[1]])
    return(grafs)
```

```python
#Imprimir
r=12
c=2
cn=0
g=graphs(lista)
fig, ax=plt.subplots(r,c,figsize=(c*10,r*10))
n=len(lista)
for i in range(r):
    if cn>23:
        cn-=1
    for j in range(c):
        ax[i][j].set_title(g[cn][1])
        ax[i][j].axis("off")
        pos=nx.spring_layout(g[cn][0])
        nx.draw_networkx(g[cn][0],pos,ax=ax[i][j],hold=False,node_size=50,with_labels=False)  
        cn+=1
```
